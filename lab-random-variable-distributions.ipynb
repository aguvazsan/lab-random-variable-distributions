```python
#importing libraries
import pandas as pd
import numpy as np
import datetime
import warnings

warnings.filterwarnings('ignore')
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from scipy.stats import mode
```

```python
#importing data
df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
```

1. Get the numerical variables from our dataset.

```python
# Check the numerical variables
numerical_df = df.select_dtypes(np.number)
```

2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.


```python
for columns in numerical_df.columns:
    sns.distplot(numerical_df[columns])
    plt.show()
```

3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.


```python
def log_transform_clean1(x):
    x = np.log(x)
    if np.isfinite(x):
        return x
    else:
        return 0

for column in numerical_df.columns:
    transformed_column = list(map(log_transform_clean1, numerical_df[column]))
    sns.distplot(transformed_column)
    plt.show()
```

Conlcusión:

La función logarítmica puede ser útil en conjuntos de datos que tienen valores positivos y tienen un rango muy amplio. Al aplicar la función logarítmica a estos conjuntos de datos, se puede reducir el rango de valores y, a su vez, disminuir la varianza y la desviación estándar.

Sin embargo, es importante tener en cuenta que la función logarítmica no es adecuada para conjuntos de datos que contienen valores nulos, ceros o valores negativos. En estos casos, la aplicación de la función logarítmica podría llevar a errores o resultados incorrectos.
